Show that for suitable constants a, b, and $\epsilon$ , the function $ f(n) = 2^{\lceil lgn \rceil }$ satisfies all
the conditions in case 3 of the master theorem except the regularity condition.

## Solución

El Teorema Maestro se aplica a recurrencias de la forma:

$
T(n) = aT\left(\frac{n}{b}\right) + f(n)
$

donde:
- $(a \geq 1$),
- $(b > 1$),
- $(f(n)$) es una función dada.

El **caso 3** del Teorema Maestro se aplica cuando se cumplen las siguientes condiciones:

1. Comparación entre $(f(n)$) y $(n^{\log_b a}$):  
   Definimos $(d = \log_b a$). El caso 3 ocurre cuando $(f(n) = \Omega(n^d))$, es decir, $(f(n)$) crece asintóticamente más rápido que $(n^d$).

3. Condición de regularidad:  
   La condición de regularidad requiere que $(af\left(\frac{n}{b}\right) \leq cf(n)\)$ para alguna constante $(c < 1$) y para $(n$) suficientemente grande.

### Paso 1: Expresar \(f(n)\)

Nos dan la función $(f(n) = 2^{\lceil \log n \rceil}$).

La función techo afecta a $(f(n)$), pero para el análisis asintótico podemos aproximarla como:

$
f(n) \approx 2^{\log n} = n
$

Así, $(f(n)$) crece aproximadamente de manera lineal, es decir, $(f(n) \in \Theta(n))$.

### Paso 2: Condición del Caso 3 (comprobar \(f(n) = \Omega(n^d)\))

El caso 3 del Teorema Maestro se aplica si:

$
f(n) = \Omega(n^d)
$

Dado que $(f(n) \in \Theta(n))$, tenemos que $(f(n) = \Omega(n))$. Para que la recurrencia del Teorema Maestro sea aplicable, necesitamos que $(n^d)$ sea menor que $(f(n))$, lo que implica que:

$
n^d = n^{\log_b a}
$

Por lo tanto, necesitamos que $(d = \log_b a < 1)$, lo que ocurre cuando $(a < b)$.

De esta manera, $(f(n) = \Omega(n^d))$ se satisface siempre que $(a < b)$.

### Paso 3: Condición de regularidad

La condición de regularidad requiere que:

$
af\left(\frac{n}{b}\right) \leq cf(n)
$

para alguna constante \(c < 1\). Consideremos:

$
af\left(\frac{n}{b}\right) = a \cdot 2^{\lceil \log \frac{n}{b} \rceil}
$

Aproximando la función techo, obtenemos:

$
af\left(\frac{n}{b}\right) \approx a \cdot \frac{n}{b}
$

Para que se cumpla la condición de regularidad, necesitamos:

$
a \cdot \frac{n}{b} \leq c \cdot n
$

o equivalentemente:

$
a \leq cb
$

Esto implica que $a$ debe ser suficientemente menor que $b$. Sin embargo, en muchos casos (especialmente cuando $a$ es cercano a $b$, esta condición puede no cumplirse, lo que viola la condición de regularidad.


La función $(f(n) = 2^{\lceil \log n \rceil}$) satisface la condición de crecimiento $(f(n) = \Omega(n^d)$) para el caso 3 del Teorema Maestro cuando $(a < b$). Sin embargo, no satisface la condición de regularidad, ya que $(af\left(\frac{n}{b}\right)$) no crece más lento que $(f(n)$) para todos los valores de $a$ y $b$.
