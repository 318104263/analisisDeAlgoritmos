# Enunciado
Usa un árbol de recursión para justificar una buena suposición de la solución de la recurrencia $T(n)=T(\alpha n)+T((1-\alpha)n)+\Theta(n)$, donde $\alpha$ es una constante en el rango $0<\alpha<1$.

# Solución


La rama más larga del árbol dependerá del valor de $\alpha$, es decir:

* Si $\alpha \geq 1-\alpha$ el algoritmo se detendrá cuando $\alpha^kn=1 \ \Rightarrow \ n = (\frac{1}{\alpha})^k \ \Rightarrow \ \log_{1/\alpha}n=k$
* Si $1-\alpha>\alpha$ el algoritmo se detendrá cuando $(1-\alpha)^kn=1 \ \Rightarrow \ n = (\frac{1}{1-\alpha})^k \ \Rightarrow \ \log_{1/1-\alpha}n=k$

Dado que la altura del árbol es logarítmica y en cada nivel se realiza un trabajo $\Theta(n)$, se tiene que $T(n)=\Theta(n)\Theta(\log n)=\Theta(n\log n)$.
