# Enunciado
Considera el procedimiento SUM-ARRAY que se muestra a continuación. Este calcula la suma de $n$ números en el arreglo $A[1:n]$. Establece un loop invariante para este procedimiento, y usa sus propiedades de inicialización, mantenimiento y terminación para mostrar que el procedimiento SUM-ARRAY retorna la suma de los números en $A[1:n]$.


# Solución
## Inicialización
Primero, para $n=0$, no existe ningún elemento que sumar y el algoritmo no realiza ninguna iteración, así que tendremos que el valor de $sum=0$, por tanto es correcto para $n=0$. 

## Mantenimiento (Inducción)
### Caso base
Supongamos que $n=1$, entonces se tendrá un sólo elemento y el ciclo for se ejecutará una única vez, en la que se le asignará a la variable $sum$ el valor $0+A[1]$, es decir, contendrá el valor del único elemento del arrelgo. Por tanto el algoritmo es correcto para $n=1$.

### Hipótesis

Ahora supongamos que el algoritmo es correcto para toda $k \geq 2$, con $k < n $, es decir, que para la iteración $k$ se cumple que se almacena el valor de la suma de los los números $A[1:k]$.

### Inducción 

Para la siguiente iteración $i=k+1$ la variable $sum$ tendrá el valor de los primeros $k$ elementos y en la línea 3 se le reasignará ese valor más el valor en $A[k+1]$, almacenando la suma de los primeros $k+1$ números, lo cual es correcto. Finalmente en la última iteración $i=n$ almacenará la suma de los números $A[1:n]$.

## Terminación
El algoritmo termina en la línea $4$, regresando la suma de los números $A[1:n]$ almacenada en la variable $sum$.
