# Enunciado
Demuestra el Teorema 3.1.



# Solución
Por demostrar $f(n)=\Theta(g(n)) \Leftrightarrow f(n) = O(g(n)) \wedge f(n) = \Omega(g(n))$. Es decir:



1. $f(n)=\Theta(g(n)) \Rightarrow f(n) = O(g(n)) \wedge f(n) = \Omega(g(n))$

Por hipótesis tenemos que $f(n)=\Theta(g(n))$, es decir,
$$
f(n) \in \{f(n) \ | \ \exists \ n_0,c_1,c_2\text{ constantes absolutas positivas, tal que } 0 \leq c_1g(n) \leq f(n) \leq c_2g(n), \ \forall n \geq n_0\}
$$
por definición. Entonces se cumple que $\exists \ c_1, c_2$ tal que $0 \leq c_1g(n) \leq f(n)$ y $0 \leq f(n) \leq c_2g(n)$. Esto quiere decir que 
$$
f(n) \in \{f(n) \ | \ \exists \ n_0,c_1\text{ constantes absolutas positivas, tal que } 0 \leq c_1g(n) \leq f(n), \ \forall n \geq n_0\}
$$
y
$$
f(n) \in \{f(n) \ | \ \exists \ n_0,c_2\text{ constantes absolutas positivas, tal que } 0  \leq f(n) \leq c_2g(n), \ \forall n \geq n_0\}
$$
Es decir, $f(n) = O(g(n)) \wedge f(n) = \Omega(g(n))$, por definición. 

2. $f(n) = O(g(n)) \wedge f(n) = \Omega(g(n)) \Rightarrow f(n)=\Theta(g(n)) $

Por hipótesis tenemos que $f(n) = O(g(n)) \wedge f(n) = \Omega(g(n))$, es decir:

$$
f(n) \in \{f(n) \ | \ \exists \ n_\Omega,c_1\text{ constantes absolutas positivas, tal que } 0 \leq c_1g(n) \leq f(n), \ \forall n \geq n_\Omega\}
$$
y
$$
f(n) \in \{f(n) \ | \ \exists \ n_O,c_2\text{ constantes absolutas positivas, tal que } 0  \leq f(n) \leq c_2g(n), \ \forall n \geq n_O\}
$$
Tomamos $n_0 = max\{n_\Theta,n_O \}$. Entonces se cumple que
$$
0 \leq c_1g(n) \leq f(n), \ \forall n \geq n_0 \quad \wedge \quad 0  \leq f(n) \leq c_2g(n), \ \forall n \geq n_0 
$$
$$
\Rightarrow 0 \leq c_1g(n) \leq f(n) \leq c_2g(n), \ \forall n \geq n_0
$$
$$
\Rightarrow f(n) \in \{f(n) \ | \ \exists \ n_0,c_1,c_2\text{ constantes absolutas positivas, tal que } 0 \leq c_1g(n) \leq f(n) \leq c_2g(n), \ \forall n \geq n_0\}
$$
$$
\therefore f(n)=\Theta(g(n))
$$
Q.E.D.
